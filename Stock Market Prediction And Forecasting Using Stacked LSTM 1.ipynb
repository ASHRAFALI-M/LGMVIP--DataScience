### Preprocess the data

```python
sc = MinMaxScaler(feature_range=(0, 1))
scaled_data = sc.fit_transform(data['Close'].values.reshape(-1, 1))
```

### Split the data into training and testing sets

```python
train_size = int(len(scaled_data) * 0.8)
train_data = scaled_data[:train_size]
test_data = scaled_data[train_size:]
```

 Define the function to create input and output sequences

```python
def create_sequences(data, seq_length):
    x = []
    y = []
    for i in range(len(data) - seq_length):
        x.append(data[i:i+seq_length])
        y.append(data[i+seq_length])
    return np.array(x), np.array(y)
```

```python
sequence_length = 20
```

Create the training sequences

```python
x_train, y_train = create_sequences(train_data, sequence_length)
```

```python
x_train
```

```python
y_train
```

 Create the testing sequences

```python
x_test, y_test = create_sequences(test_data, sequence_length)
```

```python
x_test
```

```python
y_test
```

### Build the stacked LSTM model

```python
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(x_train.shape[1], 1)))
model.add(LSTM(units=50, return_sequences=True))
model.add(LSTM(units=50))
model.add(Dense(units=1))
```

Compile the model

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

### Train the model

```python
model.fit(x_train, y_train, epochs=30, batch_size=32)
```

Make predictions

```python
predictions = model.predict(x_test)
```

### Evaluate the model

```python
predictions = sc.inverse_transform(predictions)
```

```python
rmse = np.sqrt(np.mean((predictions - sc.inverse_transform(y_test))**2))
print('Root Mean Squared Error:', rmse)
```
